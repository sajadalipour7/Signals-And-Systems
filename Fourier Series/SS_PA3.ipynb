```python
import scipy.io.wavfile
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_ct_signal(t, x_t):
    plt.plot(t,x_t)
    plt.grid()
    pass
def plot_dt_signal(n, x_n):
    index=0
    for i in n:
        tmpx=[i,i]
        tmpy=[0,x_n[index]]
        index+=1
        plt.plot(tmpx,tmpy,c='tab:blue')
    plt.scatter(n,x_n)
    plt.grid()
    pass
def find_period(data):
    maximum=data[0]
    for i in range(0,len(data)):
        if data[i]>maximum:
            maximum=data[i]
    first_time=-1
    second_time=-1
    for i in range(0,len(data)):
        if abs(data[i]-maximum)<0.001:
            if first_time==-1:
                first_time=i
            else:
                second_time=i
                break
    return second_time-first_time
def find_bk(k,data,T):
    s=0
    for i in range(0,T):
        s+=data[i]*np.cos((2*k*i*np.pi)/T)
    return s/T
def find_ck(k,data,T):
    s=0
    for i in range(0,T):
        s+=data[i]*np.sin((2*k*i*np.pi)/T)
    return s/T
def find_ak(bk,ck):
    return ((bk**2)+(ck**2))**0.5 
```

```python
# 'wave1.wav'

samplerate,data=scipy.io.wavfile.read("wave1.wav")
T=find_period(data)
length=data.shape[0]/samplerate
time=np.linspace(0.,length,data.shape[0])
plt.plot(time[:500],data[:500])
```

```python
ak=[]
bk=[]
ck=[]
xaxis=[]
for i in range(0,11):
    xaxis.append(i)
    bi=find_bk(i,data,T)
    ci=find_ck(i,data,T)
    ai=find_ak(bi,ci)
    bk.append(bi)
    ck.append(ci)
    ak.append(ai)
print("Bk :")
plt.stem(xaxis,bk)
plt.show()
print("Ck :")
plt.stem(xaxis,ck)
plt.show()
print("Ak :")
plt.stem(xaxis,ak)
plt.show()
```
